```python
from scipy import ndimage
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# c nh 
image_path = 'C:/Users/Admin/Documents/baitap TGMT/img/01/3.png'
image = Image.open(image_path)
plt.imshow(image)
```

```python
#nh xm
gray_image = image.convert('L')
#chuyn nh xm v mng
image_a = np.array(gray_image)
gradient_x = ndimage.sobel(image_a, axis=0, mode="constant")
#hin th nh
fig , axis = plt.subplots(2,1, figsize=(10,8))
plt.subplot(2,1,1)
plt.imshow(gray_image, cmap = 'gray')
plt.title("nh xm gc")

plt.subplot(2,1,2)
plt.imshow(gradient_x, cmap = 'gray')
plt.title("nh gradient gc")
```

```python
#nh xm
gray_image = image.convert('L')
#chuyn nh xm v mng
image_a = np.array(gray_image)
# Tnh gradient theo hng x (o hm ring theo x) bng mt n Prewitt x
gradient_x = ndimage.convolve(image_a, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))

# Tnh gradient theo hng y (o hm ring theo y) bng mt n Prewitt y
gradient_y = ndimage.convolve(image_a, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

# Tnh bin  gradient
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# Hin th nh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('nh xm gc')

plt.subplot(2,1,2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('nh xm gc')

plt.show()
```
