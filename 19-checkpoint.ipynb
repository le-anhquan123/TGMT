```python
from scipy import ndimage
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# đọc ảnh 
image_path = 'C:/Users/Admin/Documents/baitap TGMT/img/01/67.png'
image = Image.open(image_path)
display(image)
```

```python
#chuyển sang ảnh xám
gray_image = image.convert('L')
image_array = np.array(gray_image)
#chuyển sang ảnh nhị phân
binary_image = 1*(image_array<200)

#đánh nhãn
labeled_image, num_features = ndimage.label(binary_image)

# Tạo lưới subplot với 3 dòng và 1 cột
fig, axs = plt.subplots(3, 1, figsize=(50, 50))

plt.subplot(3,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Ảnh xám")

plt.subplot(3,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Ảnh nhị phân")

plt.subplot(3,1,3)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Đã đánh nhãn ({num_features} đối tượng)")

plt.show();
```

```python

# đọc ảnh 
image_path = 'C:/Users/Admin/Documents/baitap TGMT/img/01/8.jpg'
image = Image.open(image_path)
# Chuyển sang ảnh xám
gray_image = image.convert('L')

image_array = np.array(gray_image)

# Chuyển sang ảnh nhị phân
binary_image = 1*(image_array<200)

# Thực hiện opening để loại bỏ nhiễu và nối các đối tượng
opened_image = ndimage.binary_opening(binary_image, structure=np.ones((5,5)))

# Đánh nhãn
labeled_image, num_features = ndimage.label(opened_image)

# Tạo lưới subplot với 3 dòng và 1 cột
fig, axs = plt.subplots(4, 1, figsize=(50, 50))

plt.subplot(4,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("Ảnh xám")

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("Ảnh nhị phân")
                                      
plt.subplot(4,1,3)
plt.imshow(opened_image, cmap='gray')
plt.title("Ảnh sau opening")


plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap='gray')
plt.title(f"Đã đánh nhãn ({num_features} đối tượng)")
plt.show()
```
