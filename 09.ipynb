```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
my_path = 'C:/Users/Admin/Documents/baitap TGMT/img/01/67.png'
#dc v hin th nh 
img = Image.open(my_path)
#display(img)
plt.imshow(img)
```

```python
#chuyn sang nh xm 
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
```

```python
img_array = np.array(img_gray)
print(img_array)
```

```python
# To mt hnh v mi
plt.figure()
plt.hist(img_array.flatten(), bins=128)
plt.title('Biu  Histogram ca nh')
plt.xlabel('Gi tr ca pixel')
plt.ylabel('S lng')
plt.show()
```

```python

# Lm ni bt ng bin
plt.figure()
plt.gray()
plt.contour(img_gray, origin="image")
plt.show()
```
