```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
my_path = 'C:/Users/Admin/Documents/baitap TGMT/img/01/67.png'
#dọc và hiển thị ảnh 
img = Image.open(my_path)
#display(img)
plt.imshow(img)
```

```python
#chuyển sang ảnh xám 
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
```

```python
img_array = np.array(img_gray)
print(img_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(img_array.flatten(), bins=128)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python

# Làm nổi bật đường biên
plt.figure()
plt.gray()
plt.contour(img_gray, origin="image")
plt.show()
```

```python
# Đường dẫn ảnh
my_path = 'C:/Users/Admin/Documents/baitap TGMT/img/01/67.png'

# Đọc ảnh và hiển thị ảnh với trục tọa độ
img = Image.open(my_path).convert("L")

# Hiển thị ảnh trên trục tọa độ
plt.imshow(img)

# Tạo một ình vẽ mới
plt.figure()

# Đặt màu sắc là xám
plt.gray()

# Vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(img, origin="image")

# Đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis('equal')

# hiển thị ảnh
plt.show()
```
